```python
model = keras.Sequential(layers=[keras.layers.Dense(units=4, input_shape =[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
xs = np.array([1.0,2.0,-1.0,0.0,4.0,3.0])
ys = np.array([1.0,3.99,1.0,0.0,16.01,8.99])

```

```python
model.fit(x=xs, y=ys, epochs=500)
model.predict([2.0])
```
